```python
url = "https://boardgamegeek.com/browse/boardgame/page/1?sort=rank&sortdir=asc"


headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'
}

url = "https://boardgamegeek.com/browse/boardgame/page/1?sort=rank&sortdir=asc"
driver = webdriver.Chrome()
driver.maximize_window()
driver.get(url)
WebDriverWait(driver, 3)
#response = requests.get(url, "lxml", headers=headers)
soup = BeautifulSoup(driver.page_source, "lxml")
#html = response.content 



def item_extract(elem, selector):
    parse = elem.find_all(selector)
    if parse == []:
        return np.nan
    else:
        return parse[0]

    
link_selector = "p.smallefont.dull" 
link_lst = []

for j in range(1, 11):
    url = f"https://boardgamegeek.com/browse/boardgame/page/{j}?sort=rank&sortdir=asc"
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')
    
    for c in range(1,101):
        container = soup.select("td#CEcell_objectname{}.collection_objectname.browse")
        for i in container:
            link = item_extract(i,link_selector )
            link_lst.append(link)

        count = 0   
        d_lst=[]
        for i in link_lst:
            if link_lst[count]:
                d_lst.append(i)
            else:
                d_lst.append(np.nan)


```

```python
board_links = []

for i in range(1, 11):
    url = f"https://boardgamegeek.com/browse/boardgame/page/{i}?sort=rank&sortdir=asc"
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')
    board_links.extend([link.get('href') for link in soup.find_all('a', href=True) if link['href'].startswith("/boardgame/")])

b_links = []
for i in board_links:
    if i not in b_links:
        b_links.append(i)
        
boardgame_links = []   

unique_numbers = {}

for link in b_links:    
    middle_number = link.split('/')[2]

    if middle_number not in unique_numbers:
        unique_numbers[middle_number] = True
        boardgame_links.append(link)
len(boardgame_links)
```

```python

driver = webdriver.Chrome()
driver.maximize_window()    
    

modifucation_lst = []
for i in boardgame_links:
    url = f"https://boardgamegeek.com"+i
    driver.get(url)
    WebDriverWait(driver, 3)
    soup = BeautifulSoup(driver.page_source, "lxml")
    
    # Agree to cookies
    agree_button_selector = "button#fc-button.fc-cta-consent.fc-primary-button"
    try:
        agree_button = driver.find_element(By.CSS_SELECTOR, agree_button_selector)
        agree_button.click()
    except:
        pass
    
    selector = "li.gameplay-item" 
    # Scroll to the bottom of the page
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    
    container = soup.select("div.gameplay-item-primary")
    for i in container:
        mod = item_extract(i,selector )
        modifucation_lst.append(mod)
    
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from bs4 import BeautifulSoup

driver = webdriver.Chrome()
driver.maximize_window()  

# URL of the board game page
url = "https://boardgamegeek.com/boardgame/224517/brass-birmingham"
driver.get(url)

# Wait for the page to load
WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.CSS_SELECTOR, ".game-header-title-info")))

# Agree to cookies if the button exists
agree_button_selector = "button#fc-button.fc-cta-consent.fc-primary-button"
try:
    agree_button = driver.find_element(By.CSS_SELECTOR, agree_button_selector)
    agree_button.click()
except:
    pass

# Get the page source and parse it with BeautifulSoup
soup = BeautifulSoup(driver.page_source, "html.parser")

# Find elements containing the required information
players_element = soup.select_one(".gameplay-item .gameplay-item-primary")


# Extract text from elements
players = players_element.get_text(strip=True) if players_element else "N/A"


# Print the extracted information
print("Number of Players:", players)


driver.quit()  # Don't forget to quit the driver when done
```
